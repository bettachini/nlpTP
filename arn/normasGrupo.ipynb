# Clanificación de normas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import csv file with column names
dataFileName = 'clasificaciónNormas.csv' 
df = pd.read_csv(dataFileName, names=['Norma', 'Objetivo'])
df
```

```python
# same as above but with yrange up to 30 
df['Objetivo'].value_counts(normalize=True).mul(100).plot(kind='bar', ylabel='Proporción [%]')
plt.ylim(0, 35)
for i, v in enumerate(df['Objetivo'].value_counts(normalize=True).mul(100)):
    plt.text(i, v + 1, f'{v:.1f}', ha='center', va='bottom', rotation=90)
# title
plt.title('Clasificación de Normas ARN')
```

```python
# generate a color_map accounting for the different types found with df['Objetivo'].unique()

colour_map = {
    'Instalaciones clase 1' : 'lightblue',
    'Autorización específica' : 'lightgreen',
    'Entrenamiento de personal' : 'salmon',
    'Licenciamiento' : 'lightyellow',
    'Emplazamiento' : 'lightpink',
    'Básica seguridad radiológica' : 'lightgray',
    'Gestión de residuos' : 'lightcoral',
    'Seguridad física' : 'lightcyan',
    'No desviación de materiales nucleares' : 'lightgoldenrodyellow',
    'Transporte (adoptada)' : 'lightseagreen',
    'Gestión' : 'lightsteelblue',
    'Protección radiológica' : 'cornflowerblue',
    'Seguridad' : 'cadetblue',
    'Desmantelamiento' : 'blue',
    'Criterio de diseño' : 'darkblue',
    'Documentación' : 'darkcyan',
    'Puesta en marcha' : 'darkgoldenrod',
    'Comunicación' : 'darkgray',
    'Permiso individual' : 'darkgreen',
}
```

```python
# same graph as above but the bars are colored according to the colour_map
df['Objetivo'].value_counts(normalize=True).mul(100).plot(kind='bar', ylabel='Proporción [%]', color=[colour_map[i] for i in df['Objetivo'].unique()])

# title
plt.title('Clasificación de Normas ARN')

# ylimit
plt.ylim(0, 35)

# legend
# plt.legend(colour_map, loc='center left', bbox_to_anchor=(1, 0.5))

# percentage values on top of bars
for i, v in enumerate(df['Objetivo'].value_counts(normalize=True).mul(100)):
    plt.text(i, v + 1, f'{v:.1f}', ha='center', va='bottom', rotation=90)

plt.show()
```
