# Toma texto de PDFs

```python
# bibliotecas

import os
try:
  import PyPDF2
except ImportError:
  ! pip install PyPDF2
  import PyPDF2
```

```python
files_dir = 'normativaGeneral_descarga'
output_dir = 'scrappedPDFs'

def pdf_to_text(files_dir):
    pdf_file = open(files_dir, 'rb')
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    text = ''
    for page_num in range(pdf_reader.numPages):
        page = pdf_reader.getPage(page_num)
        text += page.extract_text()
    pdf_file.close()
    return text

def save_text_to_file(text, pdf_path):
    filename = os.path.basename(pdf_path).replace('.pdf', '.txt')
    with open('scrappedPDFs/' + filename, 'w') as text_file:
        text_file.write(text)

def main():
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    for filename in os.listdir(files_path):
        if '.pdf' in filename:
            pdf_path = os.path.join(files_path, filename)
            text = pdf_to_text(pdf_path)
            save_text_to_file(text, pdf_path)

if __name__ == '__main__':
    main()
```
